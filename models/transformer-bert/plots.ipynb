```python
import re
import matplotlib.pyplot as plt
import numpy as np

train_losses = []
val_losses = []
with open('./transformer_bert_loss.txt', 'r') as f:
    l = f.readline()
    pattern = r'^(?:Train|Valid).+loss.+(\d\.\d+).+$'
    while len(l) > 0:
        s = re.search(pattern, l)
        if s is not None:
            loss = s.group(1)
            if l.startswith('Train'):
                train_losses.append(float(loss))
            else:
                val_losses.append(float(loss)) 
        l = f.readline()

x = np.arange(len(train_losses)) + 1

plt.plot(x, train_losses, label = "Train") 
plt.plot(x, val_losses, label = "Validation") 

plt.legend() 
plt.show()
```
